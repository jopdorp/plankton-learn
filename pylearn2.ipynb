```python
import os.path
import pylearn2

from pylearn2.utils import serial
from pylearn2.datasets import dense_design_matrix
from pylearn2.datasets import preprocessing

import plankton_utils as pu
from multi_layer_perceptron import MLPClassifier

from skimage import morphology
from skimage import measure
from skimage.io import imread
from skimage.transform import resize

import itertools as it
from sklearn.preprocessing import LabelBinarizer
import numpy as np
from pylab import cm
from matplotlib import pyplot as plt
# make graphics inline
%matplotlib inline

from sklearn.neural_network import BernoulliRBM
from sklearn import linear_model, metrics
from sklearn.cross_validation import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

from sklearn import svm
```

```python
images = pu.scale_images(pu.images(pu.train_image_paths()), 25, 25)
classes = dict((cls, i) for i, cls in enumerate(pu.get_classes()))
X_ = [(classes[cls], image.ravel()) for cls, image in images]
```

```python
X = np.array([x for y, x in X_])
y = np.array([y for y, x in X_])

X_train, X_test, Y_train, Y_test = train_test_split(X, y,
                                                    test_size=0.80,
                                                    random_state=0)

print "train", X_train.shape
print "test", X_test.shape
```

```python
rbm = BernoulliRBM(random_state=0, verbose=True, learning_rate=0.01, n_iter=7, n_components=1024)
rbm.fit(X_train, Y_train)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, Y_train)
rf.score(X_test, Y_test)
```

```python
rbm = BernoulliRBM(random_state=0, verbose=True, learning_rate=0.01, n_iter=7, n_components=1024)
logistic = svm.LinearSVC();
classifier = Pipeline(steps=[('rbm', rbm), ('logistic', logistic)])
classifier.fit(X_train, Y_train)
```

```python
classifier.score(X_test, Y_test)
```

```python
rbm = BernoulliRBM(random_state=0, verbose=True, learning_rate=0.01, n_iter=7, n_components=1024)
logistic = linear_model.LogisticRegression(C=60000.0, penalty='l1')
classifier = Pipeline(steps=[('rbm', rbm), ('logistic', logistic)])
classifier.fit(X_train, Y_train)
```

```python
classifier.score(X_test, Y_test)
```

```python
rbm = BernoulliRBM(random_state=0, verbose=True, learning_rate=0.01, n_iter=7, n_components=1024)
mlp = MLPClassifier(loss='cross_entropy', output_layer='softmax', n_hidden=1000, lr = 0.03,batch_size=100)
classifier = Pipeline(steps=[('rbm', rbm), ('mlp', mlp)])
classifier.fit(X_train, Y_train)
```

```python
classifier.score(X_test, Y_test)
```

```python
view_converter = dense_design_matrix.DefaultViewConverter((32, 32, 1),('b', 0, 1, 'c'))
ddm = dense_design_matrix.DenseDesignMatrix(X=X, y=y, view_converter=view_converter)
X.shape
```
