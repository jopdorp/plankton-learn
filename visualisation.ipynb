```python
cooc_as_matrix = co_occurence_matrix.as_matrix().copy()

np.fill_diagonal(cooc_as_matrix, 0)
#cooc_as_matrix[cooc_as_matrix < 10] = 0
column_maxes = cooc_as_matrix.max(axis=0)
#alpha = cooc_as_matrix / (column_maxes * 1.0)
alpha = cooc_as_matrix / (cooc_as_matrix.max() * 1.0)

#alpha = np.zeros_like(cooc_as_matrix)
#alpha[cooc_as_matrix > 0] = 1

classes = dict([(v,k) for k,v in pu.get_classes().items()])
names = [classes[id] + str(id) for id in co_occurence_matrix.columns]
xnames = names * len(names)
ynames = list(np.array(xnames).reshape(len(names), len(names)).T.flatten())

source = ColumnDataSource(
    data=dict(
        alphas=alpha.flatten(),
        count=cooc_as_matrix.flatten(),
        xname=xnames,
        yname=ynames,
    )
)

output_file("les_mis.html")

p = figure(title="Les Mis Occurrences",
    x_axis_location="above", tools="resize,hover,save",
    x_range=list(reversed(names)),
    y_range=names)
p.plot_width = 800
p.plot_height = 800

p.rect('xname', 'yname', 0.9, 0.9, source=source, alpha='alphas', line_color=None)

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = np.pi/3

hover = p.select(dict(type=HoverTool))
hover.tooltips = OrderedDict([
    ('trueclass', '@xname'),
    ('predicted class' , '@yname'),
    ('count', '@count'),
])

show(p)      # show the plot
```
